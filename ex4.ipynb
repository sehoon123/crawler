## seaborn 라이브러리의 titanic 데이터를 로드하여 아래 질문에 맞게 시각화하시오.

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

#### 사망자와 생존자의 비율을 pie 차트로 시각화 하시오

```python
import matplotlib.pyplot as plt

# Create a new DataFrame with the count of deaths and survivors
df = titanic.groupby(['survived']).size().reset_index(name='count')

# Rename the columns
df.columns = ['Survived', 'Count']

# Plot the data in a pie chart
plt.pie(df['Count'], labels=['Dead', 'Survived'], autopct='%1.1f%%')
plt.title('Ratio of Deaths to Survivors')
plt.show()
```

#### 각 등급별 생존자와 사망자의 수를 시각화 하시오

```python
# Create a new DataFrame with the count of deaths and survivors by class
df = titanic.groupby(['class', 'survived']).size().reset_index(name='count')

# Plot the data in a bar chart
sns.barplot(x='class', y='count', hue='survived', data=df)
plt.title('Number of Survivors and Deaths by Class')
plt.ylabel('Count')
plt.xlabel('Class')
plt.show()
```
